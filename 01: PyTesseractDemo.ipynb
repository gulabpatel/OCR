```python
try:
    from PIL import Image
except ImportError:
    import Image
import pytesseract
```

```python
import matplotlib.pyplot as plt
import cv2
```

```python
im = cv2.imread("/content/Invoice.jpg", cv2.IMREAD_COLOR)
```

#Example 1

```python
plt.rcParams["figure.figsize"] = (30,30)
plt.imshow(im)

plt.show()
```

```python
# Simple image to string
print(pytesseract.image_to_string(Image.open("/content/Invoice.jpg")))
```

```python
# Get verbose data including boxes, confidences, line and page numbers
print(pytesseract.image_to_data(Image.open("/content/Invoice.jpg")))
```

#Example 2

```python
im = cv2.imread("/content/Invoice1.jpg", cv2.IMREAD_COLOR)
plt.rcParams["figure.figsize"] = (30,30)
plt.imshow(im)

plt.show()
```

```python
# Simple image to string
print(pytesseract.image_to_string(Image.open("/content/Invoice1.jpg")))
```

```python
# Get verbose data including boxes, confidences, line and page numbers
print(pytesseract.image_to_data(Image.open("/content/Invoice1.jpg")))
```

#Example 3

```python
# Get a searchable PDF
pdf = pytesseract.image_to_pdf_or_hocr('/content/Invoice2.jpg', extension='pdf')
with open('test.pdf', 'w+b') as f:
    f.write(pdf) # pdf type is bytes by default
```
